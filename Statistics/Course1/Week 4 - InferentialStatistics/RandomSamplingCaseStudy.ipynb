```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Guessed parameters to replicate SRS and analyze the sampling distribution
totPopSize = 5000

meanRegStud = 155
sdRegStud = 5

meanGymGoers = 180
sdGymGoers = 5
gymPerc = 0.3

distributionGymGoers = np.random.normal ( meanGymGoers, sdGymGoers, int(totPopSize * gymPerc) )
distributionRegStud = np.random.normal ( meanRegStud, sdGymGoers, int(totPopSize * (1 - gymPerc) ) )
distributionStudents = np.append(distributionGymGoers, distributionRegStud)

fig, (ax1, ax2, ax3 ) = plt.subplots (3, 1, figsize = (15, 15), sharex = False)

plt.xlim ([140,200])

sns.distplot(distributionGymGoers, ax = ax1)
ax1.set_title("Distribution of Body Mass of Gym Goers", fontsize = 20)
ax1.axvline(np.mean(distributionGymGoers), color = 'black')

sns.distplot(distributionRegStud, ax = ax2)
ax2.set_title("Distribution of Body Mass of Regular Students", fontsize = 20)
ax2.axvline(np.mean(distributionRegStud), color = 'black')

sns.distplot(distributionStudents, ax = ax3)
ax3.set_title("Distribution of Body Mass Overall Students", fontsize = 20)
ax3.axvline(np.mean(distributionStudents), color = 'black')

fig.tight_layout()

```

```python
#The true parameters values are established above. Now let's randomly sample from the whole population.

replication = 5000
sampleSize = 50

#first, one must create and empty array of a given shape and type

meanDistribution = np.empty(replication)
for i in range(replication):
    randomStudents = np.random.choice(distributionStudents, size = sampleSize)
    meanDistribution[i] = np.mean(randomStudents)

fig, (ax1, ax2) = plt.subplots (2, 1, figsize = (15, 15), sharex = True)

plt.xlim ([140,200])

sns.distplot(distributionStudents, ax = ax1)
ax1.set_title("Distribution of Body Mass of Overall Students", fontsize = 20)
ax1.axvline(np.mean(distributionStudents), color = 'black')

sns.distplot(meanDistribution, ax = ax2)
ax2string = "Distribution of " + str(replication) + " estimated means\n from a random sample of " + str(sampleSize) + " students"
ax2.set_title(ax2string, fontsize = 20)
ax2.axvline(np.mean(distributionStudents), color = 'blue')
ax2.axvline(np.mean(meanDistribution), color = 'black')
```

We can see that even though there's a rough variation at the mean, on average the sampling distribution has an estimated parameter value that equals the true parameter value. We see below how random samples are distributed and how they compare to the true population.

```python
graphs = (ax1, ax2, ax3, ax4, ax5, ax6) 
for i in range(1,5):
    print(graphs[i])
```

```python
fig, (ax1, ax2, ax3, ax4, ax5, ax6) = plt.subplots (6, 1, figsize = (15, 15), sharex = True)
graphs = (ax1, ax2, ax3, ax4, ax5, ax6) 

plt.xlim ([140,200])

sns.distplot(distributionStudents, ax = ax1)
ax1.set_title("Distribution of Body Mass of Overall Students", fontsize = 20)
ax1.axvline(np.mean(distributionStudents), color = 'black')

for i in range(1,6):
    randomStudents = np.random.choice(distributionStudents, size = sampleSize)
    for j in graphs:
        g = sns.distplot(randomStudents, ax = graphs[i])
        g.set_title("Distribution of Body Mass of randomly sampled Students", fontsize = 20)
        g.axvline(np.mean(distributionStudents), color = 'blue')

fig.tight_layout(pad = 2.0)
    

```

The shape of the distribution changes between different sampling events, but not the mean.

## Now, what happens if we took a biased sample?

```python
#The true parameters values are established above. Now let's randomly sample from the whole population.

replication = 5000
sampleSize = 10

#first, one must create and empty array of a given shape and type

meanDistribution = np.empty(replication)
for i in range(replication):
    #we'll look only at the distribution of gym goers
    randomStudents = np.random.choice(distributionGymGoers, size = sampleSize)
    meanDistribution[i] = np.mean(randomStudents)

fig, (ax1, ax2) = plt.subplots (2, 1, figsize = (15, 15), sharex = True)

plt.xlim ([140,200])

sns.distplot(distributionStudents, ax = ax1)
ax1.set_title("Distribution of Body Mass of Overall Students", fontsize = 20)
ax1.axvline(np.mean(distributionStudents), color = 'black')

sns.distplot(meanDistribution, ax = ax2)
ax2string = "Distribution of " + str(replication) + " means estimated from a SRS of " + str(sampleSize) + " Gym goers students"
ax2.set_title(ax2string, fontsize = 20)
ax2.axvline(np.mean(distributionStudents), color = 'black')
ax2.axvline(np.mean(meanDistribution), color = 'blue')
```
