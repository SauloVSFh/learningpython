```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('waterqualitysp.csv.csv', sep = ';')
```

```python
data.columns
data2 = data[['pH25º', 'Temp. (ºC)', 'Cond (microS)',
       'TDS (cond) (mg/L)', 'TDS (mg/L)', 'HCO3-', 'CO3-2', 'SO4-2', 'Cl-',
       'NO3-', 'F-', 'PO4-3', 'Br-', 'Ca+2', 'Mg+2', 'Na+', 'K+', 'Ba+2',
       'Sr+2', 'SiO2']]
data2.head()
```

```python
datalong = pd.melt(frame = data, id_vars = ['Identificação', 'CIDADE', 'UF'], value_vars = data2.columns)
```

```python
datalong
```

```python
databox = datalong.loc[~(datalong.variable.isin(['pH25º', 'Temp. (ºC)', 'Cond (microS)', 'TDS (cond) (mg/L)','TDS (mg/L)']))]
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(y = 'value', x = 'variable', data = databox)
```
